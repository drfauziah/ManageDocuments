## Mengganti Halaman

Section ini akan membahas cara cepat untuk mengganti halaman tertentu di beberapa dokumen PDF menggunakan python

1. import library yang dibutuhkan : os, glob, PyPDF2, PdfFileReader, PdfFileWriter

```python
import os
import glob
from PyPDF2 import PdfFileWriter, PdfFileReader
```

2. Memanggil file yang berisi halaman baru

```python
halaman_baru = PdfFileReader("halaman_baru.pdf") ## edit nama file pdf
```

3. List multiple pdf yang akan di replace, pastikan berurutan seperti urutan di halaman_baru.pdf (bisa menggunakan fungsi sorted()). Jika halaman yang akan di replace adalah sama untuk semua dokumen, maka tidak perlu berurutan. 

```python
list_pdf = glob.glob("Folder/*.pdf") ## edit path folder/file
```

4. Replace page

```python
for i in range(len(list_pdf)):
    file_baru = PdfFileWriter()
    pdf = PdfFileReader(list_pdf[i])
    for j in range(pdf.getNumPages()):
        if j == pdf.getNumPages() - 2:        ## -2 berarti mengganti halaman ke-2 dari halaman terakhir, edit lokasi halaman yang akan diganti
            file_baru.addPage(halaman_baru.getPage(i)) ## halaman ke-i bisa diedit tergantung keperluan
        else:
            file_baru.addPage(pdf.getPage(j))
            
    file_baru.write(open(list_pdf[i], "wb")) ## menyimpan file yang telah di replace dengan nama dan lokasi file yang sama dari file sebelumnya
```
