### Menggabungkan beberapa PDF

Section ini membahas cara menggabungkan beberapa file pdf menjadi satu file pdf

1. **Merger** : Halaman terakhir dari file pertama dilanjutkan dengan halaman pertama dari file kedua dan seterusnya.

```python
from PyPDF2 import PdfFileMerger

pdfs = ['file_pertama.pdf', 
        'file kedua.pdf',
        'file_ketiga.pdf']

merger = PdfFileMerger()

for pdf in pdfs:
    merger.append(pdf)

merger.write("SML_PJB_ROLLOUT_PEMBANGUNAN_SOKET_LAPORAN PELAKSANAAN TRAINING SOE_RBG.pdf")
merger.close()
```

2. **InsertPage** : Menambahkan halaman tertentu dari file pertama.

      Berikut adalah contoh untuk menambahkan file ttd ke halaman ke-2 dari semua file dalam list_pdf

```python
import os
import glob
from PyPDF2 import PdfFileWriter, PdfFileReader

#list file
list_pdf = glob.glob("Folder/*.pdf")
ttd = PdfFileReader("Lembar_tanda_tangan.pdf")

for i in range(len(list_pdf)):
    kosong = PdfFileWriter()
    reader = PdfFileReader(list_pdf[i])
    for j in range(reader.getNumPages()):
        kosong.addPage(reader.getPage(j))
    kosong.insertPage(ttd.getPage(0), index=-1)
            
    kosong.write(open(list_pdf[i], "wb"))
```
