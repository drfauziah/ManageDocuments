#### Membuat list jumlah halaman file pdf dari folder tertentu

```python
import pandas as pd
import os
from PyPDF2 import PdfFileReader
df = pd.DataFrame(columns=['fileName', 'fileLocation', 'pageNumber'])
for root, dirs, files in os.walk(r'E:\Folder\*.pdf'): ## edit path
    for f in files:
        if f.endswith(".pdf"):
            pdf=PdfFileReader(open(os.path.join(root, f),'rb'))
            df2 = pd.DataFrame([[f, os.path.join(root,f), pdf.getNumPages()]], columns=['fileName', 'fileLocation', 'pageNumber'])
            df = df.append(df2, ignore_index=True)
```

- List lokasi file pdf, nama file, dan jumlah halaman dari setiap file

```python
df[['fileName', 'fileLocation', 'pageNumber']]
```

- Menghitung total halaman dari semua file pdf

```python
print("total halaman pdf dalam folder =", df['pageNumber'].sum())
```
